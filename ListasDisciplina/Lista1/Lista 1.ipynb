**Questão 3)** Implemente um aplicativo que receba o tamanho n de um filtro
espacial com dimensões n × n, e a matriz com os pesos do filtro, e aplique o
filtro numa imagem, gerando e salvando a imagem resultante no disco.

```python
import cv2
import numpy
from matplotlib import pyplot
```

```python
img_before = cv2.imread('./first_frame.jpg')

pyplot.imshow(img_before)
pyplot.title('Imagem antes')
pyplot.show()
```

```python
def cross_correlation2d(source_img, size_filter):
    # Ler a imagem o size do filtro
    def kernel(size_filter):
        kernel_lst = []
        for i in range(0, size_filter):
            for j in range(0, size_filter):
                print("Peso do elemento:", i, j)
                kernel_lst.append(float(input("")))
        kernel_img = numpy.array(kernel_lst, dtype=float)
        return kernel_img.reshape(n,n)
    
    k = kernel(size_filter)
    
    result_img = numpy.zeros_like(source_img)
    
    for i in range(0, source_img.shape[0]):
        for j in range(0, source_img.shape[1]):
            sum_ = 0
            for u in range(0, size_filter):
                for v in range(0, size_filter):
                    sum_ += k[u][v] * source_img[(i) - u][(j) - v]
            result_img[i][j] = sum_
            
    return result_img
```

```python
n = int(input("Digite o tamanho do filtro:"))

img_after = cross_correlation2d(img_before, n)

pyplot.subplot(121), pyplot.imshow(img_before), pyplot.title('Antes do kernel')
pyplot.subplot(122), pyplot.imshow(img_after), pyplot.title('Após o kernel')

pyplot.gcf().set_size_inches((15, 15))
pyplot.show()
```

**Questão 9** Um filtro passa-baixa bastante conhecido 
é o filtro Gaussiano. Através de uma máscara determinada 
por um desvio padrão σ, é possı́vel fazer o que chamamos 
de suavização gaussiana de uma imagem. Considere o filtro 
5x5 de desvio padrão unitário abaixo:

g = $\begin{bmatrix}
1/273 & 4/273 & 7/273 & 4/273 & 1/273 \\
4/273 & 16/273 & 26/273 & 16/273 & 4/273 \\
7/273 & 26/273 & 41/273 & 26/273 & 7/273 \\
4/273 & 16/273 & 26/273 & 16/273 & 4/273 \\
1/273 & 4/273 & 7/273 & 4/273 & 1/273 \\
\end{bmatrix}$

Crie uma imagem I de dimensões 100 × 100 que contém um único 
valor não nulo localizado no centro imagem. 
Utilizando o filtro gaussiano, calcule: g * I, I * g, 
g $\otimes$ I e I $\otimes$ g.

```python
import cv2
import numpy
from random import uniform
from matplotlib import pyplot
```

```python
img = numpy.ones((100,100,3), numpy.float32) * 0.0
for i in range(0, 3):
    img[50][50][i] = 1
pyplot.imshow(img)
pyplot.title('Original')
pyplot.show()
```

```python
gauss_kernel = numpy.array([[1/273, 4/273, 7/273, 4/273, 1/273],
                           [4/273, 16/273, 26/273, 16/273, 4/273],
                           [7/273, 26/273, 41/273, 26/273, 7/273],
                           [4/273, 16/273, 26/273, 16/273, 4/273],
                           [1/273, 4/273, 7/273, 4/273, 1/273]])
```

```python
g_cc_I = cv2.filter2D(gauss_kernel, -1, img) # g * I
I_cc_g = cv2.filter2D(img, -1, gauss_kernel) # I * g

pyplot.subplot(131), pyplot.imshow(img), pyplot.title('Original')
#pyplot.xticks([]), pyplot.yticks([])
pyplot.subplot(132), pyplot.imshow(g_cc_I), pyplot.title('g $\otimes$ I')
#pyplot.xticks([]), pyplot.yticks([])
pyplot.subplot(133), pyplot.imshow(I_cc_g), pyplot.title('I $\otimes$ g')
#pyplot.xticks([]), pyplot.yticks([])
pyplot.gcf().set_size_inches((10, 10))
pyplot.show()
```

```python
def g_conv_I_function(source_img, conv_filter):
    k = conv_filter.shape[0]
    
    result_img = numpy.zeros_like(source_img)
    
    for i in range(0, source_img.shape[0]):
        for j in range(0, source_img.shape[1]):
            conv_sum = 0
            for u in range(0, k):
                for v in range(0, k):
                    conv_sum += conv_filter[u][v] * source_img[(i) - u][(j) - v]
            result_img[i][j] = conv_sum
            
    return result_img
```

```python
from scipy import signal

#g_conv_I = signal.convolve2d(gauss_kernel, img, mode='full') # g * I
I_conv_g = g_conv_I_function(img, gauss_kernel) # I * g

pyplot.subplot(131), pyplot.imshow(img), pyplot.title('Original')
pyplot.subplot(132), pyplot.imshow(g_conv_I), pyplot.title('g * I')
pyplot.subplot(133), pyplot.imshow(I_conv_g), pyplot.title('I * g')
pyplot.gcf().set_size_inches((10, 10))
pyplot.show()
```
