A seguir definimos o algoritmo de remoção de background:

```python
def remove_background(frame, background_frame, threshold, beta, background_mask=None):
    '''
    Dada a imagem de background (background_frame) e um limiar (threshold)
    retorna a imagem de entrada (frame) sem o background.
    '''
    if background_mask is None:
        
        background_frame = cv2.cvtColor(background_frame, cv2.COLOR_RGB2GRAY)
        foreground_frame = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
                
        adjust_brightness = numpy.vectorize(lambda x: 255 if x + beta > 255 else x + beta)
        foreground_frame = adjust_brightness(foreground_frame).astype('uint8')
        
        background_threshold = numpy.vectorize(lambda x: 0 if x < threshold else 1)
        background_mask = background_threshold(foreground_frame - background_frame)
    
    result_frame = frame.copy()

    result_frame[:,:,0] = numpy.multiply(result_frame[:,:,0], background_mask)
    result_frame[:,:,1] = numpy.multiply(result_frame[:,:,1], background_mask)
    result_frame[:,:,2] = numpy.multiply(result_frame[:,:,2], background_mask)
    
    return result_frame, background_mask
```

Aplicando o algoritmo nas imagens dadas:

```python
THRESHOLD, BETA = 60, 30

removed_background_frame, background_mask = remove_background(frame=second_frame, background_frame=first_frame, threshold=THRESHOLD, beta=BETA)

pyplot.imshow(removed_background_frame)
pyplot.title('Removed Background')
pyplot.show()
```

```python
def zeroes_pad(array, pad_size=1):
    def pad_with(vector, pad_width, iaxis, kwargs):
        pad_value = kwargs.get('padder', 0)
        vector[:pad_width[0]] = pad_value
        vector[-pad_width[1]:] = pad_value
    return numpy.pad(array, pad_size, pad_with)
```

```python
def convolution(source_img, conv_filter):
    k = conv_filter.shape[0]
    
    result_img = numpy.zeros_like(source_img)
    source_img = zeroes_pad(source_img, pad_size=k//2)
    
    for i in range(k//2, source_img.shape[0] - k//2):
        for j in range(k//2, source_img.shape[1] - k//2):
            conv_sum = 0
            for u in range(0, k):
                for v in range(0, k):
                    conv_sum += conv_filter[u][v] * source_img[(i + k//2) - u][(j + k//2) - v]
            result_img[i-k//2][j-k//2] = conv_sum
            
    return result_img

def convolve_3d(source_img, conv_filter):

    source_img[:,:,0] = convolution(source_img[:,:,0], conv_filter)
    source_img[:,:,1] = convolution(source_img[:,:,1], conv_filter)
    source_img[:,:,2] = convolution(source_img[:,:,2], conv_filter)
    
    return source_img
```

4. Implemente dois tipos de filtro para remoção de ruídos e encontre exemplos onde cada um deles funciona melhor.

```python
def add_noise(image, noise_str):
    
    # https://stackoverflow.com/questions/22937589/how-to-add-noise-gaussian-salt-and-pepper-etc-to-image-in-python-with-opencv
    
    if noise_str == 'salt_and_pepper':
        row, col, cha = image.shape
        s_vs_p = 0.5
        amount = 0.2
        out = numpy.copy(image)
          
        # Salt mode
        num_salt = numpy.ceil(amount * image.size * s_vs_p)
        coords = [numpy.random.randint(0, i - 1, int(num_salt)) for i in image.shape]
        out[tuple(coords)] = 1
        
        # Pepper mode
        num_pepper = numpy.ceil(amount * image.size * (1. - s_vs_p))
        coords = [numpy.random.randint(0, i - 1, int(num_pepper)) for i in image.shape]
        out[tuple(coords)] = 0
        
        return out
    elif noise_str == 'gauss':
        row, col, cha = image.shape
        
        mean = 0
        var = 10
        sigma = var ** 0.5
        
        gauss = numpy.random.normal(mean, sigma, (row, col, cha))
        gauss = gauss.reshape(row, col, cha)
        noisy = image + gauss
        
        return noisy.astype('uint8')
    else:
        return 0
```

```python
def high_pass(k):
    hp_filter = (-1 / k**2) * numpy.ones(shape=(k, k))
    hp_filter[k//2][k//2] = (k**2 - 1) / k**2
    return hp_filter
```

```python
for image_file in os.listdir('./images'):
    img_path = os.path.join('./images', image_file)
    
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    for n in range(3, 12, 4):

        dest_img = convolve_3d(source_img=img, conv_filter=high_pass(k=n))
        pyplot.imshow(dest_img)
        pyplot.title(f'Kernel Size {n}x{n}')

        pyplot.show()
```
